```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'train' is your DataFrame from previous cells
d = train.select_dtypes(include=np.number).apply(pd.value_counts).T.sum(axis=1)
# Selects numerical features, counts values, transposes and sums to get total count per feature

dT=d.T
dT.plot.bar(y='count')
plt.title("Bar plot of the count of numeric features",
          fontsize=17)
plt.xlabel("Numeric features",fontsize=17)
plt.ylabel("Count",fontsize=17)
```
