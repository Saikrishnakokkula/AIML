```python
import matplotlib.pyplot as plt
x=[1,2,4,7,10]
y=[3,6,8,20,40]
z=[10,20,30,40,50]
a=[1,2,3,4,5]
b=[2,4,6,8,10]
c=[3,6,9,12,15]
fig=plt.figure(figsize=(5,5))
ax=fig.add_subplot(111,projection='3d')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
plt.plot(x,x,color='red')
plt.plot(x,y,color='green')
plt.plot(x,z,color='blue')
plt.plot(y,a,color='orange')
plt.plot(y,x,color='violet')
plt.plot(y,c,color='darkgreen')
plt.plot(z,z,color='skyblue')
plt.plot(z,a,color='black')

plt.title("3D-FIGURE")
plt.show()
```
